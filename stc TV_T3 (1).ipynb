```python
from surprise import Reader

# Define the format of the input data
reader = Reader(rating_scale=(1, 5))
```

```python
import pandas as pd
from surprise import Dataset, SVD


# Build the Recommender System using SVD algorithm
data = Dataset.load_from_df(dataframe[['user_id_maped', 'program_name', 'rating']], reader)
trainset = data.build_full_trainset()
model = SVD()
model.fit(trainset)

# Get the top 5 recommendations for all users who watched "Moana"
moana_users = dataframe[dataframe['program_name'] == 'Moana']['user_id_maped'].tolist()
all_moana_recs = []
for user_id in moana_users:
    # Predict the ratings for all the programs for the current user
    user_ratings = [(program_name, model.predict(user_id, program_name).est) for program_name in dataframe['program_name'].unique()]
    # Sort the ratings in descending order and get the top 5 recommendations for the current user
    user_recs = sorted(user_ratings, key=lambda x: x[1], reverse=True)[:5]
    all_moana_recs += user_recs

# Remove duplicates from the list of recommendations
all_moana_recs = list(set(all_moana_recs))

# Print out the top 5 recommendations
print("Top 5 Recommendations for all users who watched 'Moana'")
for program_name, rating in all_moana_recs[:5]:
    print(program_name, rating)
```
